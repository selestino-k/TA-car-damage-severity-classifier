```python
import pandas as pd
import matplotlib.pyplot as plt

csv_files = ['baselinecnn_training_log.csv', 'vgg19_training_log.csv', 'resnet50_training_log.csv', 'densenet121_training_log.csv', 'inceptionv3_training_log.csv']
dataframes = [pd.read_csv(file) for file in csv_files]
model_labels = ['Baseline CNN', 'VGG19', 'ResNet50', 'DenseNet121','InceptionV3']
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
colors = ['b', 'g', 'r', 'm', 'y']
linestyles = '-'

#Loss
for i, df in enumerate(dataframes):
    axs[0, 0].plot(df['epoch'], df['loss'], label=model_labels[i], color=colors[i], linestyle=linestyles)
axs[0, 0].set_title('Training Loss')
axs[0, 0].set_xlabel('Epoch')
axs[0, 0].set_ylabel('Training Loss')
axs[0, 0].legend()

# Accuracy
for i, df in enumerate(dataframes):
    axs[0, 1].plot(df['epoch'], df['accuracy'], label=model_labels[i], color=colors[i], linestyle=linestyles)
axs[0, 1].set_title('Training Accuracy')
axs[0, 1].set_xlabel('Epoch')
axs[0, 1].set_ylabel('Training Accuracy')
axs[0, 1].legend()

#Validation Loss
for i, df in enumerate(dataframes):
    axs[1, 0].plot(df['epoch'], df['val_loss'], label=model_labels[i], color=colors[i], linestyle=linestyles)
axs[1, 0].set_title('Validation Loss')
axs[1, 0].set_xlabel('Epoch')
axs[1, 0].set_ylabel('Validation Loss')
axs[1, 0].legend()

#Validation Accuracy
for i, df in enumerate(dataframes):
    axs[1, 1].plot(df['epoch'], df['val_accuracy'], label=model_labels[i], color=colors[i], linestyle=linestyles)
axs[1, 1].set_title('Validation Accuracy')
axs[1, 1].set_xlabel('Epoch')
axs[1, 1].set_ylabel('Validation Accuracy')
axs[1, 1].legend()

plt.tight_layout()
plt.show()
```
